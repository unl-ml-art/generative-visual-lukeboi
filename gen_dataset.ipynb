```python
import csv
import numpy as np
import cv2

lines = []

with open('output_SRTMGL3.asc', 'r') as fd:
    reader = csv.reader(fd)
    for row in reader:
#         print(row[0].split(" "))
        l = row[0].split(" ")
        l = list(filter(None, l))
#         print(l)
        lines.append(l)

lines_np = np.array(lines, dtype=float)
print(lines_np.shape)


lines_np -= lines_np.min()
lines_np *= 255.0 / lines_np.max()

output_image_size = 64
n_written = 0
for i in range(len(lines_np) - output_image_size):
    for j in range(len(lines_np[0]) - output_image_size):
        img = lines_np[i:i + output_image_size, j:j + output_image_size]

        if img.max() - img.min() < 50:
            continue
        
#         img -= img.min()
#         img *= 255.0 / img.max()
        
        n_written += 1
#         filename = "subset/" + str(i) + "_" + str(j) + ".png"
        filename = "subset/classa/" + str(n_written) + ".png"
        cv2.imwrite(filename, img)
        
        if n_written > 50000:
            break
    print(i, "th row,", n_written, "files written")
    if n_written > 50000: 
        break
```
